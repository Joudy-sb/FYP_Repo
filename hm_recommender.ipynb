```python
import pandas as pd

# Load the articles dataset
customers_df = pd.read_csv("customers.csv")

# Display the first few rows
customers_df.head()
```

```python
# Count the number of filtered customers
num_filtered_customers = filtered_customers.shape[0]

# Print the result
print(f"Number of active customers aged 20-25: {num_filtered_customers}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

# Load the annotated articles dataset
articles_annotated_df = pd.read_csv("articles_annotated.csv")

# Function to get unique purchased items and display images
def display_customer_purchases(customer_id, transactions_df, articles_annotated_df, num_images=10):
    """
    Extract unique items purchased by a given customer and display their images.

    Parameters:
        customer_id (str): The ID of the customer
        transactions_df (DataFrame): Transactions dataset
        articles_annotated_df (DataFrame): Articles metadata dataset
        num_images (int): Number of images to display
    """
    # Get all transactions for the customer
    customer_transactions = transactions_df[transactions_df["customer_id"] == customer_id]

    # Merge transactions with article metadata
    customer_items = customer_transactions.merge(
        articles_annotated_df[["article_id", "folder", "item_name"]], 
        on="article_id", 
        how="left"
    )

    # Keep only unique items
    customer_unique_items = customer_items.drop_duplicates(subset=["article_id"]).copy()

    # Construct image paths
    customer_unique_items["image_path"] = customer_unique_items.apply(
        lambda row: os.path.join("images", f"0{row['folder']}", row["item_name"]), axis=1
    )

    # Limit the number of images to display
    num_images = min(num_images, len(customer_unique_items))

    # Create subplots
    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))

    # Ensure axes is always iterable
    if num_images == 1:
        axes = [axes]  # Convert single Axes object to a list

    # Loop through and display images
    for i, (_, row) in enumerate(customer_unique_items.head(num_images).iterrows()):
        image_path = row["image_path"]

        if os.path.exists(image_path):  # Check if the image exists
            img = Image.open(image_path)
            axes[i].imshow(img)
            axes[i].set_title(f"Item {row['article_id']}")
            axes[i].axis("off")
        else:
            print(f"Image not found: {image_path}")

    plt.show()

# Extract the top 3 customer IDs
top_customer_ids = top_customers_details["customer_id"].tolist()

# Loop through customers 1, 2, and 3 and display their purchased items
for i, customer_id in enumerate(top_customer_ids[0:3], start=1):  # Include customer 1
    print(f"\nDisplaying purchases for Customer {i}: {customer_id}\n")
    display_customer_purchases(customer_id, transactions_df, articles_annotated_df, num_images=10)
```
